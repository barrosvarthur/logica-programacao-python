Atividade 01 – Lógica Programação Python




```
# 1) Crie um programa que escreva 'Meu Primeiro Programa!' na tela.

print("Meu primeiro programa!")
```

```
# 2) Faça um programa que leia o nome de uma pessoa e mostre uma mensagem de boas-vindas.

name = input("Escreva seu nome: ")
print(f"Bem vindo, {name}!")
```

```
# 3) Crie um programa que leia dois números e mostre a soma entre eles.

number1 = float(input("Insira o primeiro número a ser somado: "))
number2 = float(input("Insira o segundo número a ser somado: "))

print(f"A soma de {number1} + {number2} = {number1 + number2}.")
```

```
# 4) Faça um programa que leia um número inteiro e mostre na tela o seu sucessor e seu antecessor.

number = int(input("Insira um número para ver seu antecessor e sucessor: "))

print(f"Antecessor: {number - 1}\nSucessor: {number + 1}")
```

```
# 5) Crie um algoritmo que leia um número e mostre o seu dobro, triplo e raiz quadrada.

number = int(input("Insira um número: "))

print(f"Dobro = {number*2}\nTriplo = {number*3}\nRaiz quadrada = {number**0.5}")
```

```
# 6) Desenvolva um programa que leia as duas notas de um aluno, calcule e mostre a sua média.

note1 = float(input("Insira sua nota da 1ª prova: "))
note2 = float(input("Insira sua nota da 2ª prova: "))

print(f"Sua média é {(note1 + note2)/2}")
```

```
# 7) Escreva um programa que leia um valor em metros e o exiba convertido em centímetros e milímetros.

meters = float(input("Insira um valor em metros "))

print(f"Centímetros: {meters*100}")
print(f"Milímetros: {meters*1000}")
```

```
# 8) Faça um algoritmo que leia o preço de um produto e mostre seu novo preço, com 5% de desconto.

price = float(input("Insira o preço para calcular o desconto: "))

print(f"O valor  com desconto fica: R$ {price * 0.95}")
```

```
# 9) Faça um algoritmo que leia o salário de um funcionário e mostre seu novo salário, com 15% de aumento.

wage = float(input("Insira o salário para receber o aumento: R$ "))

print(f"O salário com o aumento de 15% fica: R$ {wage *1.15}")
```

```
# 10) Escreva um programa que converta uma temperatura digitando em graus Celsius e converta para graus Fahrenheit.

temperature = float(input("Insira uma temperatura em celsius para converter em fahrenteit: "))

print(f"A temperatura em fahrenteit é: {(temperature * 1.8) + 32}")
```
